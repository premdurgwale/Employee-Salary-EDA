```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/employee_salary_dataset.csv")
```

```python
plt.figure(figsize=[10,8])
plt.hist(df['Experience_Years'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Experience_Years')
plt.show()
```

```python
plt.figure(figsize=[10,8])
plt.boxplot(df['Monthly_Salary'].dropna(), vert=True)
plt.title("Monthly_Salary Boxplot")
plt.ylabel("Monthly_Salary")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Correlation Heatmap', fontdict={'fontsize':36, 'fontweight':'bold'})
plt.title("Correlation Heatmap")
sns.heatmap(df.corr(numeric_only=True), annot=True, linewidths=0.5, cmap='Blues')
plt.show()
```

**Education_Level Pie Chart**

```python
plt.figure(figsize=[16,6])
plt.title("Pie Chart",fontdict={'fontsize':30,'fontweight':'bold'})
plt.pie(df['Education_Level'].value_counts(sort=False),
        labels=df['Education_Level'].unique(),
        autopct='%1.2f%%',
        startangle=90,
        #explode=[0,0,0]
        )
plt.show()
```

**Avereage Salary by Department**

```python
plt.figure(figsize=[18,8])
plt.title('Bar Plot\n', fontdict={'fontsize':36, 'fontweight': 'bold'})
sns.barplot(x='Department', y='Monthly_Salary', data=df,ci=None)
plt.show()
```

```python
plt.figure(figsize=[8,4])
plt.title('Hist plot\n', fontdict={'fontsize':36,'fontweight':'bold'})
sns.histplot(x='Age',data=df,kde=True)
```

```python
plt.figure(figsize=[16,6])
plt.title("Pie Chart",fontdict={'fontsize':30,'fontweight':'bold'})
plt.pie(df['Gender'].value_counts(sort=False),
        labels=df['Gender'].unique(),
        autopct='%1.2f%%',
        startangle=90,
        #explode=[0,0,0]
        )
plt.show()
```

**Distribution of departments across **citys**.**

```python
plt.figure(figsize=[14,8])
plt.title('Scatter plot\n', fontdict={'fontsize': 36, 'fontweight': 'bold'})
sns.scatterplot(x='City', y='Department', data=df, s = 150)
```

**Average Salary by City**

```python
plt.figure(figsize=[18,8])
plt.title('Bar Plot\n', fontdict={'fontsize':36, 'fontweight': 'bold'})
sns.barplot(x='City', y='Monthly_Salary', data=df,ci=None)
plt.show()
```
